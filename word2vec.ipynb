```python
from gensim.test.utils import datapath
from gensim import utils
```

```python
class MyCorpus(object):
    """An interator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = 'ALL_reviews.txt'
        with open(corpus_path, encoding='utf-8') as corpus:
            for line in corpus:
                # assume there's one document per line, tokens separated by whitespace
                yield utils.simple_preprocess(line)
```

```python
sentences = MyCorpus()
model = gensim.models.Word2Vec(sentences=sentences)
```

```python
import numpy as np


x_train = []

for s in sentences:
    avg_vec = np.zeros(100)
    words_cnt = len(s)
    for token in s:
        if token in model.wv.vocab:
            vw = model.wv[token]
            avg_vec = np.add(avg_vec, vw)
    avg_vec = avg_vec / words_cnt    
    
    x_train.append(avg_vec)
```

```python
# Оценки для отзывов
y_train = []

for i, row in excel_file.iterrows():
    if (row['title'] != 'Криминальное чтиво' and
        row['title'] != 'Маленькая Мисс Счастье' and
        row['title'] != 'Амели'):
        y_train.append(row['label'])
```

```python
len(y_train)
```

```python
class MyTestCorpus(object):
    """An interator that yields sentences (lists of str)."""

    def __iter__(self):
        corpus_path = 'all_my_reviews.txt'
        with open(corpus_path, encoding='utf-8') as corpus:
            for line in corpus:
                # assume there's one document per line, tokens separated by whitespace
                yield utils.simple_preprocess(line)

my_test_sentences = MyTestCorpus()
```

```python
x_test = []

for s in my_test_sentences:
    avg_vec = np.zeros(100)
    words_cnt = len(s)
    for token in s:
        if token in model.wv.vocab:
            vw = model.wv[token]
            avg_vec = np.add(avg_vec, vw)
    avg_vec = avg_vec / words_cnt    
    
    x_test.append(avg_vec)
    
# Оценки для отзывов
y_test = []

for i, row in excel_file.iterrows():
    if (row['title'] == 'Криминальное чтиво' or
        row['title'] == 'Маленькая Мисс Счастье' or
        row['title'] == 'Амели'):
        y_test.append(row['label'])
```

```python
# Т.к. на моих тестовых наилучший показатель по F-мере был у метода RF, word2vec я буду оценивать на этом же методе

from sklearn.ensemble import RandomForestClassifier
rfclf = RandomForestClassifier(max_depth=20, random_state=0)
rfclf.fit(x_train, y_train)
```

```python
# predict the labels on validation dataset
predictions_RF = rfclf.predict(x_test)
print(predictions_RF)


# Use accuracy_score function to get the accuracy
from sklearn.metrics import accuracy_score

print("RF Accuracy Score -> ",accuracy_score(predictions_RF, y_test)*100)
```

```python
from sklearn.metrics import precision_recall_fscore_support


print(precision_recall_fscore_support(y_test, predictions_RF, average='weighted'))
```
